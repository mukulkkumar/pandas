### Sort a dataframe by its Index
* The **sort_index** method sorts the dataframe by its index positions/labels.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("Crime Prediction in Chicago_Dataset.csv")
```

```python
dataframe.columns.values
```

```python
dataframe.sort_index()
dataframe.sort_index(ascending=False)
```

### Rank values with the rank method
* The **rank** method assigns a numeric ranking to each Series value.
* Pandas will assign the same rank to equal values and create a "gap" in a dataset for the ranks

```python
dataframe['Community Area'].rank()
```

```python
dataframe['Community Area'].rank().astype(int)
```

```python
dataframe["Community Area Rank"] = dataframe['Community Area'].rank().astype(int)
```
