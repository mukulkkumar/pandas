### Fill in missing values with the fillna method
* The fillna method replaces missing NaN values with its arguement.
* The fillna method available on both DataFrame and Series.
* An extracted Series is a view on the original DataFrame, but the fillna method returns a copy.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("Crime Prediction in Chicago_Dataset.csv")
```

```python
dataframe
```

```python
dataframe['Location'] = dataframe['Location'].fillna(0)
```

```python
dataframe
```

```python
dataframe['Longitude'] = dataframe['Longitude'].fillna(value=0)
```
