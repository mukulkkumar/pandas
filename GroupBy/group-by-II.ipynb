### Method on the groupby object
* Use square bracket on the **DataFrameGroupBy** object to "extract" a column from the original **DataFrame**.
* The resulting **SeriesGroupBy** object will have aggregation methods available on it.
* Pandas will perform the calculation on every group within the collection.
* For example, the **sum** method will sum together the number column for every row by group/category.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv")
```

```python
primary_type = dataframe.groupby('Primary Type')
```

```python
primary_type["Community Area"]
```

```python
primary_type["Community Area"].sum()
```

```python
primary_type["Community Area"].mean()
```

```python
primary_type[["Community Area", "Ward"]].sum()
```
