### Iterating through groups
* The **DataFrameGroupBy** object supports the apply method (just like a **Series** and a **DataFrame** do).
* The **apply** method invokes a function on every nested **DataFrame** in the **DataFrameGroupBy** object.
* It captures the return values of the functions and collects them in a new **DataFrame** (the return value).

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv")
```

```python
primary_type = dataframe.groupby('Primary Type')
```

```python
# Find the two "Community Area" in each primary area with most area
def top_two_community_area_by_primary_area(primary_type):
    return primary_type.nlargest(2, "Community Area")

primary_type.apply(top_two_community_area_by_primary_area)
```
