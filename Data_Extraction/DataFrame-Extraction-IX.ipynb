### Create random sample with the sample method
* The **sample** method returns a specified one or more random rows from the **DataFrame**.
* Customize the **axis** parameter to extract random columns.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col="ID").sort_index()
```

```python
dataframe.head()
```

```python
dataframe.sample()
```

```python
dataframe.sample(n=5)
```

```python
dataframe.sample(n=3, axis='rows')
```

```python
dataframe.sample(n=3, axis='columns')
```
