### Retrieve row by index label with loc accessor
* The **loc** accessor retrieves one or more rows by index label.
* Provide a pair of square brackets after the accessor.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv")
```

```python
dataframe = dataframe.set_index('ID')
```

```python
dataframe.loc[12592454]
```

```python
dataframe.loc[[12592454, 12785595]]
dataframe.loc[[12785595, 12592454]]
```

```python
dataframe.loc[12592454: 12888104]
```

```python
# If the index position is in reverse, it won't give any result
dataframe.loc[12888104: 12592454]
```

```python
dataframe.loc[12827191:]
```

```python
dataframe.loc[:12888104]
```
