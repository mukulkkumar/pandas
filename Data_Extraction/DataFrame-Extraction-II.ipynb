### Retrieve rows by index position with iloc accessor
* The **iloc** accessor retrieves one or more rows by index position.
* Provide a pair of square brackets after the accessor.
* **iloc** accepts single values, list and slices.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv")
```

```python
dataframe.head()
```

```python
dataframe.iloc[2]
```

```python
dataframe.iloc[[2,5]]
```

```python
dataframe.iloc[3:5]
```

```python
dataframe.iloc[0:4]
```

```python
dataframe.iloc[:4]
```

```python
len(dataframe)
```

```python
dataframe.iloc[45420:]
```
