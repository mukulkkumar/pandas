### Apply method with dataframes
* The **apply** method invokes a function on every column or every row in the **DataFrame**.
* Pass the uninvoked function as the first argument to the apply method.
* Pass the **axis** parameter an argument of **"columns"** to invoke the function on every row.
* Pandas will pass in the row's values as a **Series** object. We can use accessors like **loc** and **iloc** to extract the column's values for that row.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col="ID").sort_index()
```

```python
dataframe.head()
```

```python
dataframe["Description"].apply(len)
```

```python
# Community Area
def get_community_specification(row):
    area = row.loc["Community Area"]

    if area > 50:
        print("Enough Space")
    else:
        print("Not Enough Space")

dataframe.apply(get_community_specification, axis='columns')
```
