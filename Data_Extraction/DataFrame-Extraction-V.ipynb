### Overwrite value in a dataframe
* Use the **iloc** or **loc** accessor on the dataframe to target a value, then provide the equal sign and a new value.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col="ID")
```

```python
# Don't overwrite the value in dataframe like this, this will give you a warning. However it will overwrite dataframe data

dataframe["Case Number"].loc[12592454] = "JF346523"
```

```python
dataframe.loc[dataframe["Case Number"] == "JF346523"]
```

```python
# The right way of overwriting the dataframe data
dataframe.loc[12592454, "Case Number"] = "JF346527"
```

```python
dataframe.loc[dataframe["Case Number"] == "JF346527"]
```
