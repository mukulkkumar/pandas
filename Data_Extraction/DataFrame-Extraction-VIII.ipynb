### Delete rows or columns from a dataframe
* The **drop** method deletes one or more rows/columns from a **DataFrame**.
* Pass the **index** or **columns** parameters a list of the column names to remove.
* The **pop** method removes and returns a single **Series** (It mutates the **DataFrame** in the process).
* Python **del** keyword also removes a single **Series**.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col="ID").sort_index()
```

```python
dataframe.head()
```

```python
df = dataframe.drop(columns=["Primary Type", "Description"])
```

```python
df.head()
```

```python
df = dataframe.drop(index=[26596, 26630])
```

```python
df.head()
```

```python
df = dataframe.drop(index=[26596, 26630], columns=["Primary Type", "Description"])
```

```python
df.head()
```

```python
dataframe.pop("Block")
```

```python
dataframe.columns.values
```

```python
del dataframe["Primary Type"]
```
