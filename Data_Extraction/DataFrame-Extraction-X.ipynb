### The nsmallest and nlargest methods
* The **nlargest** method returns a specified number of rows with the largest values from a given column.
* The **nsmallest** method returns rows with the smallest values from a given column.
* The **nlargest** and **nsmallest** methods are more efficient than sorting the entire **DataFrame**.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col="ID").sort_index()
```

```python
dataframe.head()
```

```python
# Retrieve the 4 Beat with the highest/largest box office gross
dataframe.sort_values("Beat", ascending=False).head(4)
```

```python
dataframe.nlargest(n=4, columns="Beat")
```

```python
dataframe["Beat"].nlargest(4)
```

```python
dataframe.nsmallest(4, columns="Beat")
```

```python
dataframe["Beat"].nsmallest(4)
```
