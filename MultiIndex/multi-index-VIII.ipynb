### The unstack method
* The **unstack** method moves a row index to the column index (the inverse of the stack method).
* By default, the **unstack** method will move the innermost index.
* We can customize the moved index with the **level** parameter.
* The **level** parameter accepts the level's index position or its name. It can also accept a list of positions/names.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col=["ID", "Primary Type"]).sort_index().stack()
```

```python
dataframe.head()
```

```python
dataframe.unstack()
```

```python
dataframe.unstack().columns
```

```python
dataframe.unstack(level=1)
```

```python
dataframe.unstack(level="Primary Type")
```

```python
dataframe.unstack(level=0)
```

```python
dataframe.unstack([1, 0])
```

```python
dataframe.unstack(['Primary Type', 'ID'])
```

```python
dataframe.unstack(['Primary Type', 'ID']).sort_index(axis=1)
```
