### The transpose method
* The **transpose** method inverts/flips the horizontal and vertical axes of the **DataFrame**.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col=["District", "Primary Type"]).sort_index()
```

```python
dataframe.head()
```

```python
start = (31, "BATTERY")
end = (31, "DECEPTIVE PRACTICE")
dataframe.loc[start:end]
```

```python
dataframe.loc[start:end].transpose()
```
