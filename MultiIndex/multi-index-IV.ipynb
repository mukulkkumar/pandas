### The sort_index Method on MultiIndex DataFrame
* Using the **sort_index** method, we can target all levels or specific levels of the **MultiIndex**.
* To apply a different sort order to different levels, pass a list of Booleans.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col=["District", "Primary Type"])
```

```python
dataframe.head()
```

```python
dataframe.sort_index()
```

```python
dataframe.sort_index(ascending=True)
```

```python
dataframe.sort_index(ascending=False)
```

```python
dataframe.sort_index(ascending=[False, True])
```

```python
dataframe.sort_index(ascending=[True, False])
```
