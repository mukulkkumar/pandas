### Rename index levels
* Invoke the **set_names** method on the **MultiIndex** to change one or more level names.
* Use the **names** and **level** parameter to target a nested index at a given level.
* Alternatively, pass **names** a list of strings to overwrite all level names.
* The **set_names** method returns a copy, so replace the original index to alter the **DataFrame**.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col=["District", "Primary Type"]).sort_index()
```

```python
dataframe.head()
```

```python
dataframe.index.set_names(names="Location", level=0)
```

```python
dataframe.index.set_names(names="Type", level=1)
```

```python
dataframe.index.set_names(names=["Location", "Type"])
```

```python
dataframe.index = dataframe.index.set_names(names=["Location", "Type"])
```
