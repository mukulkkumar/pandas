### The between Method
* The **between** method returns True if a **Series** value is found its range.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("Crime Prediction in Chicago_Dataset.csv")
```

```python
valid_community_area = dataframe['Community Area'].between(10, 100)
```

```python
dataframe[valid_community_area]
```

```python
dataframe.columns.values
```

```python
updated_range = dataframe['Updated On'].between('9/14/2023', '9/14/2024')
```

```python
dataframe[updated_range]
```
