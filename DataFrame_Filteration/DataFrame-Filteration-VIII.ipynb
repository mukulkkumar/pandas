### The drop_duplicates method
* The **drop_duplicates** method delete rows with duplicate values.
* By default, it will remove a row if all of its values are shared with another row.
* The **subset** parameter configures the columns to look for duplicate values within.
* Pass a list to **subset** parameter to look fr duplicates across multiple columns.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("Crime Prediction in Chicago_Dataset.csv")
```

```python
dataframe.drop_duplicates()

dataframe.drop_duplicates('Primary Type')

dataframe.drop_duplicates('Primary Type', keep="first")

dataframe.drop_duplicates('Primary Type', keep="last")

dataframe.drop_duplicates('Primary Type', keep=False)
```

```python
dataframe.drop_duplicates(["Primary Type", "Location Description"]).sort_values('Primary Type')

dataframe.drop_duplicates(["Primary Type", "Location Description"], keep="last").sort_values('Primary Type')
```
