### Filter DataFrame based on a condition
* Pandas need a **Series** of Booleans to perform a filter
* Pass the Boolean Series inside square brackets after a **DataFrame**
* We can generate a Boolean Series using a wide variety of opterations (equality, inequality, less than, greater than, inclusion e.t.c)

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("Crime Prediction in Chicago_Dataset.csv")
```

```python
dataframe["Location Description"] == "RESIDENCE"
```

```python
dataframe[dataframe["Location Description"] == "RESIDENCE"].head()
```

```python
residence = dataframe["Location Description"] == "RESIDENCE"
```

```python
dataframe[residence].head()
```

```python
dataframe[dataframe['Arrest']].head()
```

```python
dataframe[dataframe['Beat'] > 2500].head()
```

```python
dataframe[dataframe['Ward'] <= 2.0]
```

```python
dataframe[dataframe['Updated On'] > '9/14/2023 15:40']
```
