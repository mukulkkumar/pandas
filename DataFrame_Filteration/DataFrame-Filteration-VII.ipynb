### The duplicated Method
* The duplicated method returns True if a **Series** value is a duplicate
* Pandas will mark one occurence of a repeated value as a non-duplicate
* Use the **keep** parameter to designate whether the first or last occurence of a repeated value should be considered the "non-duplicate"
* Pass False to the **keep** parameter to mark all occurences of repeated values as duplicates.
* Use the tilde symbol **(~)** to invert a **Series** values. Trues will become False, and False will become trues.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("Crime Prediction in Chicago_Dataset.csv")
```

```python
dataframe['Primary Type'].duplicated()
```

```python
dataframe[dataframe['Primary Type'].duplicated()]
```

```python
dataframe[dataframe['Primary Type'].duplicated(keep="first")]
```

```python
dataframe[dataframe['Primary Type'].duplicated(keep="last")]
```

```python
dataframe[dataframe['Primary Type'].duplicated(keep=False)]
```

```python
dataframe[~dataframe['Primary Type'].duplicated(keep=False)]
```
