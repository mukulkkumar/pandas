### Filter with more than one condition (AND)
* Add the & operator in between two Boolean **Series** to filter by multiple conditions.
* We can assign the **Series** to variables to make the syntax more readable.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("Crime Prediction in Chicago_Dataset.csv")
```

```python
is_theft = dataframe['Primary Type'] == 'THEFT'
is_residence = dataframe['Location Description'] == 'RESIDENCE'
```

```python
is_theft & is_residence
```

```python
dataframe[is_theft & is_residence]
```

```python
is_arrestted = dataframe['Arrest'] == True
```

```python
dataframe[is_theft & is_residence & is_arrestted]
```
