### Methods on Index and Columns
* Use the **index** and **columns** attributes to access the **DataFrame** index/column labels.
* These objects support string methods via their own **str** attribute.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col="ID").sort_index()
```

```python
dataframe.head()
```

```python
dataframe.index = dataframe.index + 1
```

```python
dataframe.head()
```

```python
dataframe.columns = dataframe.columns.str.lower()
```
