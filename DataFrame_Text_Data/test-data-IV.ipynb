### The split method
* The **str.split** method splits a string by the occurence of a delimiter. Pandas returns a **Series** of lists.
* Use the **str.get** method to access a nested list element by its index position.
* The **expand** parameter returns a **DataFrame** instead of a **Series** of lists.
* The **n** parameter limits the number of splits.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../Crime Prediction in Chicago_Dataset.csv", index_col="ID").sort_index()
```

```python
dataframe.head()
```

```python
dataframe['Description'].str.split(" ")
```

```python
dataframe['Description'].str.split(" ").str.get(0)
```

```python
dataframe['Description'].str.split(" ", expand=True)
```

```python
dataframe['Description'].str.split(" ", expand=True, n=2)
```
